```python
def calculate_distance(LATITUDE1, LONGITUDE1, LATITUDE2, LONGITUDE2, API):
    url = f'https://dev.virtualearth.net/REST/v1/Routes/DistanceMatrix?key={API}&origins={LATITUDE1},{LONGITUDE1}&destinations={LATITUDE2},{LONGITUDE2}&travelMode=driving&distanceUnit=mi'
    response = requests.get(url)
    return response.json()
```
